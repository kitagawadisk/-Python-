```python
#1-3 H形鋼の断面係数の計算　関数
def H_katako(H, B, tw, tf): #H形鋼の断面二次モーメントと断面係数を計算する関数
    b = B - tw
    h = H - 2*tf
    Ix = ((B*H**3)/12 - (b*h**3)/12) *10**-4
    Zx = Ix/(H/2/10)
    Iy = (2*tf*B**3 + (H*tf*2)*tw**3)/12 *10**-4
    Zy = (2*tf*B**3 + (H*tf*2)*tw**3)/(6*B) *10**-3
    return(Ix, Zx, Iy, Zy) #H形鋼の断面二次モーメントと断面係数を計算する関数

def str_maker1(Ix, Zx, Iy, Zy): #断面係数の計算結果の文章をつくる関数
    s = "1-3 H形鋼の断面係数\n" # s += "" と s = s + ""　は同じ。
    s += "Ix = %.0f cm4 (鋼材表 20200 cm4)\n"%(Ix)
    s += "Zx = %.0f cm3 (鋼材表 1350 cm3)\n"%(Zx)
    s += "Iy = %.0f cm4 (鋼材表 6750 cm4)\n"%(Iy)
    s += "Zy = %.0f cm3 (鋼材表 450 cm3)"%(Zy)
    return s

#ここから。（関数１つにつき、１つの用途がよい。）
Ix13, Zx13, Iy13, Zy13 = H_katako(300,300,10,15) #H形鋼の断面二次モーメントと断面係数を計算する関数
s_13 = str_maker1(Ix13, Zx13, Iy13, Zy13) #断面係数の計算結果を表示する関数
print(s_13)
```

```python
#2-1 耐風梁 H形鋼元断面の検討
def H_katako(H, B, tw, tf): #H形鋼の断面二次モーメントと断面係数を計算する関数
    b = B - tw
    h = H - 2*tf
    Ix = ((B*H**3)/12 - (b*h**3)/12) *10**-4
    Zx = Ix/(H/2/10)
    Iy = (2*tf*B**3 + (H*tf*2)*tw**3)/12 *10**-4
    Zy = (2*tf*B**3 + (H*tf*2)*tw**3)/(6*B) *10**-3
    return(Ix, Zx, Iy, Zy) #H形鋼の断面二次モーメントと断面係数を計算する関数
    
def str_maker2(l, a, w_long, w_short): #元断面で計算結果の文章を作成する関数
    ftl = 156 #N/mm2
    fts = 235 #N/mm2
    M_long = w_long*a*l**2/8 #kNm !自重を無視
    M_short = w_short*a*l**2/8 #kNm
    s = "[耐風梁 元断面の検討]\n"
    s += "\tスパン長 L = %.2f m\n"%(l)
    s += "\t強軸方向の荷重 [負担幅 %.1f m, 荷重 %.1f kN/m2]\n"%(a,w_long)
    s += "\t弱軸方向の荷重 [負担幅 %.1f m, 荷重 %.1f kN/m2]\n"%(a,w_short)
    s += "\t弱軸方向 M（中央部の値）= %.2f kNm\n"%M_long
    s += "\t強軸方向 M（中央部の値）= %.2f kNm\n"%M_short
    s += "\tIx = %.0f cm4, Zx = %.0f cm3, Iy = %.0f cm4, Zy = %.0f cm3\n"%(Ix,Zx,Iy,Zy)
    s += "\t許容応力度 f長期 %.0f N/mm2, f短期 %.0f N/mm2\n"%(ftl, fts)
    s += "\t長期検定比 %.2f（強軸）+ %.2f（弱軸）= %.2f\n"%(0/Zx*1000, M_long/Zy*1000/ftl, 0/Zx/ftl + M_long/Zy*1000/ftl)
    s += "\t短期検定比 %.2f（強軸）+ %.2f（弱軸）= %.2f\n"%(M_short/Zx*1000/fts, M_long/Zy*1000/fts, M_short/Zx*1000/fts+ M_long/Zy*1000/fts)
    return s

#ここから（４）
l = 9.6 #スパンm
a = 3 #負担幅m
w_long = 1.3 #長期等分布荷重kN/m2
w_short = 2.6 #短期等分布荷重kN/m2
Ix, Zx, Iy, Zy = H_katako(300,300,10,15) #H形鋼のI、Z算定関数
s1 = str_maker2(l, a, w_long, w_short) #計算書の文章を作成
print(s1)
```

```python
#2-2-1 耐風梁 汎用断面の検討（関数部分）
import matplotlib.pyplot as plt #グラフを描画するライブラリ
import matplotlib.patches as patches
import copy

#凡例の場所
locs = [
    "upper left",
    "upper right",
    "lower left",
    "lower right",
    "center left",
    "center right",
    "upper center",
    "lower center",
    "center",
    "best",
    "right",
]

def edge_create(r1): #エッジの作成
    e0 = [[],[]]
    for ri in r1: #Rectang情報
        #print(ri)
        x0 = ri["center_x"] - ri["B"]/2
        y0 = ri["center_y"] - ri["H"]/2
        x1 = ri["center_x"] + ri["B"]/2
        y1 = ri["center_y"] + ri["H"]/2
        e0[0].append(x0)
        e0[1].append(y0)

        e0[0].append(x1)
        e0[1].append(y0)
        
        e0[0].append(x1)
        e0[1].append(y1)

        e0[0].append(x0)
        e0[1].append(y1)
    return(e0)

def anchor_center(r1): #基準点から中心と左下の算定
    r2 = []
    for ri in r1:
        if ri["anchor"] == "center":
            leftdown_x = ri["anchor_x"] - ri["B"]/2
            leftdown_y = ri["anchor_y"] - ri["H"]/2
            center_x = ri["anchor_x"]
            center_y = ri["anchor_y"]
        elif ri["anchor"] == "lower_left":
            leftdown_x = ri["anchor_x"]
            leftdown_y = ri["anchor_y"]
            center_x = ri["anchor_x"] + ri["B"]/2
            center_y = ri["anchor_y"] + ri["H"]/2
        elif ri["anchor"] == "upper_left":
            leftdown_x = ri["anchor_x"]
            leftdown_y = ri["anchor_y"] - ri["H"]
            center_x = ri["anchor_x"] + ri["B"]/2
            center_y = ri["anchor_y"] - ri["H"]/2
        elif ri["anchor"] == "lower_right":
            leftdown_x = ri["anchor_x"] - ri["B"]
            leftdown_y = ri["anchor_y"]
            center_x = ri["anchor_x"] - ri["B"]/2
            center_y = ri["anchor_y"] + ri["H"]/2
        elif ri["anchor"] == "upper_right":
            leftdown_x = ri["anchor_x"] - ri["B"]
            leftdown_y = ri["anchor_y"] - ri["H"]
            center_x = ri["anchor_x"] - ri["B"]/2
            center_y = ri["anchor_y"] - ri["H"]/2
        else:
            print("ANCHOR ERROR")
        ri["center_x"] = center_x #辞書にセンターを追加
        ri["center_y"] = center_y
        ri["leftdown_x"] = leftdown_x
        ri["leftdown_y"] = leftdown_y
        r2.append(ri)
    return r2

def plot_geometical(r11, c0=[0,0], save=""): #矩形情報, 図芯位置
    #断面二次モーメント:: geometrical moment of inertia
    #断面一次モーメント:: statical moment of area
    fig = plt.figure() #グラフの準備
    ax = plt.axes() #グラフ描画域１の準備
    ax.grid(True) #グリッドは一番下に配置したいのでこの位置で指定。

    centerlist = [[], []]
    for ri in r11:
        #print(leftdown_x, leftdown_y)
        #図形での確認
        centerlist[0].append(ri["center_x"])
        centerlist[1].append(ri["center_y"])
        r = patches.Rectangle(xy=(ri["leftdown_x"], ri["leftdown_y"]), width=ri["B"], height=ri["H"], ec='#000000', fill=False)
        ax.add_patch(r)
    
    #図芯位置
    ax.scatter(c0[0], c0[1], label="Figure core", s=6)
    #各矩形の重心位置リスト
    ax.scatter(centerlist[0], centerlist[1], label="Rectangle center", s=6)
    plt.axis('scaled')
    ax.set_aspect('equal')
    ax.legend(loc=locs[1], bbox_to_anchor=(0.5, 0, 0.5, 1), frameon=False)
    ax.set_xlim(-200, 200)
    ax.set_ylim(-300, 300)
    
    if save == "":
        plt.show() #描いたグラフを表示
    else:
        plt.savefig(save) #描いたグラフを保存
    plt.close() 

def XY_create(r1): #矩形のエッジ情報から図の最大エッジを取得, 
    cp0 = [[],[]]
    for ri in r1: #Rectang情報
        #print(ri)
        x0 = ri["center_x"] - ri["B"]/2
        y0 = ri["center_y"] - ri["H"]/2
        x1 = ri["center_x"] + ri["B"]/2
        y1 = ri["center_y"] + ri["H"]/2
        cp0[0].append(x0) #左下
        cp0[1].append(y0)

        cp0[0].append(x1) #右下
        cp0[1].append(y0)
        
        cp0[0].append(x1) #右上
        cp0[1].append(y1)

        cp0[0].append(x0) #左上
        cp0[1].append(y1)
    
    e0 = edge_create(r1)
    lx = max(abs(max(e0[0])),abs(min(e0[0])))
    ly = max(abs(max(e0[1])),abs(min(e0[1])))
    return(lx,ly)

class Core2: #矩形の断面二次モーメントの計算仮定を保管しておくクラス
    #そのうち計算仮定の出力にも対応したい。
    def __init__(self, r1):
        self.r1 = r1 #矩形断面の元座標データ
        self.r2 = 0 #矩形断面の図芯基準の座標データ
        self.xl1 = [] #断面1次モーメントの計算仮定の各値リスト
        self.yl1 = [] #断面1次モーメントの計算仮定の各値リスト
        self.Al = [] #各矩形断面の断面積リスト
        self.xl2 = [] #断面2次モーメントの計算仮定の各値リスト
        self.yl2 = [] #断面2次モーメントの計算仮定の各値リスト
        self.c0 = 0 #新座標の元座標に対する移動量
        self.dxl = []
        self.dyl = []

        self.create_core() #計算関数で上記変数に値を保存
        self.Iy2 = self.SUMx2/10000
        self.Zy2 = self.SUMx2/10000/(self.capY/10)
        self.Ix2 = self.SUMy2/10000
        self.Zx2 = self.SUMy2/10000/(self.capX/10)

    def create_core(self): #図芯と断面二次モーメントを算定する関数
        r1 = self.r1
        #断面一次モーメントの算定
        SUMx1 = 0
        SUMy1 = 0
        SUMA = 0
        for ri in r1:
            Seci = ri["B"]*ri["H"]
            SUMxi = Seci*ri["center_x"]
            SUMyi = Seci*ri["center_y"]
            SUMx1 += SUMxi
            SUMy1 += SUMyi
            SUMA += Seci
            self.xl1.append(SUMxi)
            self.yl1.append(SUMyi)
            self.Al.append(Seci)
            
        #図芯位置
        c0 = [SUMx1/SUMA, SUMy1/SUMA]
        #print("図芯",c0)
        self.c0 = c0

        #断面二次モーメント計算前にcenter座標を図芯基準に変更
        r2 = copy.deepcopy(r1) #これをしないとr2の変更がr1に影響した。
        for i in range(len(r2)):
            r2[i]["center_x"] = r2[i]["center_x"] - c0[0]
            r2[i]["center_y"] = r2[i]["center_y"] - c0[1]
            r2[i]["leftdown_x"] = r2[i]["leftdown_x"] - c0[0]
            r2[i]["leftdown_y"] = r2[i]["leftdown_y"] - c0[1]
        self.r2 = r2

        #断面二次モーメントの算定
        SUMx2 = 0
        SUMy2 = 0
        for ri in r2:
            SUMxi = ri["B"]*ri["H"]**3/12 + ri["B"]*ri["H"]*ri["center_y"]**2
            self.dyl.append(ri["center_y"])
            SUMx2 += SUMxi #X軸廻りの断面二次モーメント
            self.xl2.append(SUMx2)
            SUMyi = ri["B"]**3*ri["H"]/12 + ri["B"]*ri["H"]*ri["center_x"]**2
            self.dxl.append(ri["center_x"])
            SUMy2 += SUMyi #y軸廻りの断面二次モーメント
            self.yl2.append(SUMy2)
        
        self.SUMx2 = SUMx2
        self.SUMy2 = SUMy2
        self.capX,self.capY = XY_create(r2) #mm, mm
        #print(self.SUMx2, self.SUMy2,self.capX,self.capY)
```

```python
import csv

def read_csv(csvfile): #csv読込関数
    with open(csvfile, 'rt') as f: #csvを読み込む
        header = next(csv.reader(f))
        reader = csv.reader(f)
        kk = [row for row in reader]

    kk2 = list(list(zip(*kk))[0]) #nomberだけ
    kk2 = [int(i) for i in kk2] #nomberを整数型に修正
    kk2 = list(set(kk2)) #nomberの重複削除

    r2list = []
    for k in kk2: #nomber毎のループ
        r22 = []
        for i in range(len(kk)): #データをリストに入れる
            if int(kk[i][0]) == k:
                r22.append({"anchor_x": float(kk[i][1]), "anchor_y":float(kk[i][2]),"H":float(kk[i][3]),"B":float(kk[i][4]), "anchor":str(kk[i][5])})
        r2list.append(r22)
    return r2list, kk2

#ここから
r2list, kk2 = read_csv('kukeilist1.csv') #csvのデータを読み取り
r2 = r2list[0]
for j in range(len(r2list)):
    print("") #一行空白
    for i in range(len(r2list[j])): #printで確認
        print(r2list[j][i])
```

```python
#2-3 計算書のワードファイル作成例
import docx #ワードを使う場合のライブラリ
from docx.shared import Pt
from docx.oxml.ns import qn
from docx.shared import Mm                    # 単位や色が定義されているSharedクラス
from docx.enum.text import WD_ALIGN_PARAGRAPH # 段落位置の定義クラスをインポート
from docx.enum.table import WD_TABLE_ALIGNMENT

def docx_gothic(p): #計算書タイトルのフォーマット
    p.font.name = 'ＭＳ Ｐゴシック'#フォントサイズ変更
    p._element.rPr.rFonts.set(qn('w:eastAsia'), u'ＭＳ Ｐゴシック')
    p.font.size = Pt(11) #フォントサイズ変更
    p.font.bold = False #太文字とはしない。

def docx_mincho(p): #主文章のフォーマット
    p.font.name = 'ＭＳ Ｐ明朝'#フォントサイズ変更
    p._element.rPr.rFonts.set(qn('w:eastAsia'), u'ＭＳ Ｐ明朝')
    p.font.size = Pt(10.5) #フォントサイズ変更
    p.font.bold = False #太文字とはしない。

def docx_figure_center(doc, save, tx): #図にタイトルを付けて中心配置する
    tables = doc.add_table(rows=1, cols=1)
    tables.alignment = WD_TABLE_ALIGNMENT.CENTER
    p = tables.rows[0].cells[0].paragraphs[0]
    p.alignment=WD_ALIGN_PARAGRAPH.CENTER
    r = p.add_run()
    r.add_picture(save, height=Mm(60))
    r.add_break()
    r.add_text(tx)

def create_docx(save_docx):
    doc = docx.Document() #Word文書の新規作成
    #（１）タイトル
    p = doc.add_paragraph().add_run("耐風梁の補強の検討\n")
    docx_gothic(p) #ゴシックの書式設定
    #（２）図を添付して図番追加
    docx_figure_center(doc, "./img/耐風梁補強断面図.png", "図1-1 耐風梁補強断面図")
    
    for i in range(len(kk2)): #繰り返し
        #（３）改ページを追加
        doc.add_page_break() #改ページを追加
        #（４）元断面の検討文章
        p = doc.add_paragraph().add_run(s1)
        docx_mincho(p) #明朝の書式設定
        #（５）補強断面の断面二次モーメント算定図添付
        docx_figure_center(doc, savelist[i], "図1-2-%i 補強部の図芯"%(kk2[i]))
        #（６）補強断面の検討文章
        p = doc.add_paragraph().add_run(s2list[i]) #文字列s2を追加
        docx_mincho(p) #明朝の書式設定
        
    p = doc.add_paragraph("以上") #以上を記入
    p.alignment = WD_ALIGN_PARAGRAPH.RIGHT #右寄

    doc.save(save_docx)

    #参考 https://gammasoft.jp/support/how-to-use-python-docx-for-word-file/

#ここから
save_docx = "reinforcement.docx" #保存名
create_docx(save_docx) #ワードファイルを作成して保存
```

```python
#2-5 pdfを画像に変換して表示
import os
from pathlib import Path
from pdf2image import convert_from_path
poppler_path =  r"C:\soft\poppler-22.04.0\Library\bin"

def pdf_to_image(file_name): #PDFをimgに変換
    pdf_path = Path(file_name) # PDFファイルのパス
    img_path_base = "./img"
    (name, extention) = os.path.splitext(file_name)
    images = convert_from_path(pdf_path ,poppler_path=poppler_path) 
    for i in range(len(images)):
        images[i].save(img_path_base + "/" + name + '-' + str(i+1)+'.jpg', 'JPEG')

pdf_to_image("./reinforcement.pdf") 

from IPython.display import Image #画像を表示するだけ
#Image(data='./img/reinforcement-1.jpg',retina=True) #1ページ目だけjupyter-labに表示
Image(data='./img/reinforcement-2.jpg',retina=True) #2ページ目だけjupyter-labに表示
```
